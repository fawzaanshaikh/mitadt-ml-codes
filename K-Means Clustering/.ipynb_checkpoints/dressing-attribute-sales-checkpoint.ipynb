```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder # For change in the labels of the output
```

```python
df = pd.read_excel("Attribute DataSet.xlsx")
```

```python
df.head()
```

```python
le = LabelEncoder()
le.fit(df["Style"])
le.classes_

df["Style"] = le.transform(df["Style"])

km = KMeans(n_clusters=2, max_iter=400,verbose=True,tol=0.2)
pred = km.fit_predict(df[df.head()])
score = accuracy_score(df["Dress_ID"], pred)
score
```
